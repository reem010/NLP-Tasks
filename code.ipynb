```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return 'a'  # Adjective
    elif treebank_tag.startswith('V'):
        return 'v'  # Verb
    elif treebank_tag.startswith('N'):
        return 'n'  # Noun
    elif treebank_tag.startswith('R'):
        return 'r'  # Adverb
    else:
        return 'n'

def processing(doc):
    text = re.sub(r'[^a-zA-Z\s]', '', doc)
    text = text.lower()
    words = text.split()
    tagged_words = pos_tag(words)
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word, pos=get_wordnet_pos(tag)) for word, tag in tagged_words]
    # stemmer = PorterStemmer()
    # words = [stemmer.stem(word) for word in words]  # stemming
    stop_words = set(stopwords.words("english"))
    document = [word for word in words if word not in stop_words and len(word) >= 3 ]
    unique_words = set(document)

    return document, unique_words
```

```python
doc_Science, unique_words_Science = processing(generated_documents["Science"])
doc_Sport, unique_words_Sport = processing(generated_documents["Sport"])
doc_Lifestyle, unique_words_Lifestyle = processing(generated_documents["Lifestyle"])
```

```python
documents = [doc_Science, doc_Sport, doc_Lifestyle]
unique_words_documents = [
    unique_words_Science,
    unique_words_Sport,
    unique_words_Lifestyle,
]
```

# **From Scratch**

**Calculate TF**

```python
tf = [Counter(doc) for doc in documents]
tf = [{word: count  for word, count in doc.items()} for doc in tf]
```

```python
print(tf)
```

**Calculate IDF**

```python
idf = {}
for doc in unique_words_documents:
    for word in doc:
        idf[word] = idf.get(word, 0) + 1
        # if word == "proud":
        #    print(idf.get(word, 0))
idf = {
    word: math.log((len(unique_words_documents)+1) / (count + 1)) + 1
    for word, count in idf.items()
}
```

```python
print(idf)
```

**Calculate TF*IDF**

```python
tfidf = [{word: tf_word * idf[word] for word, tf_word in doc.items()} for doc in tf]
```

**Normalize TF*IDF**

```python
tfidf_normalized = []
for doc in tfidf:
    norm = math.sqrt(sum([x**2 for x in doc.values()]))
    tfidf_normalized.append({word: tf_word / norm for word, tf_word in doc.items()})
```

**Convert to DataFrame**

```python
words = list(idf.keys())
data = np.zeros((len(documents), len(words)))
for i, doc in enumerate(tfidf_normalized):
    for word, score in doc.items():
        data[i, words.index(word)] = score
scratch_results_dataframe = pd.DataFrame(data, columns=words)
```
